```python
import numpy as np
from bokeh.core.properties import value
from bokeh.io import show, output_file, output_notebook
from bokeh.models import ColumnDataSource, HoverTool, FactorRange
from bokeh.plotting import figure
from bokeh.palettes import Spectral9
from bokeh.transform import dodge
from bokeh.layouts import gridplot
```

```python
output_notebook()
dat = np.loadtxt('swcfs_prioritization_data.dat', delimiter=',')
alts = ['Ainf','Ainj','B','C','D','E','F','G','H']
metrics = ['Volume', 'Cost', 'Adtl Ben', 'Cons & Ops']
Scores = ['Low', 'Medium', 'High']
```

```python
pList = []

for scIx in xrange(3):

    curScore = scIx+1
    
    curdat = dat[ dat[:,3]==curScore, : ]
    mags = np.zeros((4,len(alts)))
    for i in range(curdat.shape[0]):
        mags[ int(curdat[i,2]-1), int(curdat[i,1]-1) ] = curdat[i,4]

    data = {'metrics'       :   metrics}
    for i in xrange(len(alts)):
        data[alts[i]] = mags[:,i]

    source = ColumnDataSource(data=data)
    p = figure(x_range=metrics, y_range=(0,1.5), plot_width=900, plot_height=300, title='{:s}-Score Alts by Metric'.format(Scores[scIx]),toolbar_location=None, tools='') 

    # positions = np.arange(-0.5,0.51,0.125)
    positions = np.arange(-0.4,0.41,0.1)
    for i in xrange(len(alts)):
        p.vbar(x=dodge('metrics', positions[i], range=p.x_range), top=alts[i], width=0.08, source=source, 
               color=Spectral9[i], muted_color=Spectral9[i], muted_alpha=0.1, legend=value(alts[i]), muted=True)

    p.x_range.range_padding = 0.1
    p.xgrid.grid_line_color = None
    
    p.legend.location = "top_center"
    p.legend.orientation = "horizontal"
    p.legend.click_policy="mute"
    
    pList.append(p)
```

```python
## grid = gridplot( [[pList[0], None, None], [None, pList[1], None], [None, None, pList[2]]] )
grid = gridplot([[pList[2]],[pList[1]],[pList[0]]])

show(grid)
```
